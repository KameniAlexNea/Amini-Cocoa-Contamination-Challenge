```python
# Load all CSV files into a single DataFrame
all_data = pd.concat([pd.read_csv(file).tail(1) for file in files], ignore_index=True)

# Filter the DataFrame to include only the specified columns
columns_of_interest = [
    "metrics/precision(B)",
    "metrics/recall(B)",
    "metrics/mAP50(B)",
    "metrics/mAP50-95(B)",
    "val/box_loss",
    "val/cls_loss",
    "val/dfl_loss",
    "epoch",
    "timestamp",
    "checkpoint_dir_name",
    "done",
    "training_iteration",
    "trial_id",
    "date",
    "time_this_iter_s",
    "time_total_s",
    "pid",
    "hostname",
    "node_ip",
    "time_since_restore",
    "iterations_since_restore",
]
filtered_data = all_data[columns_of_interest]

# Sort the DataFrame to find the top 10 experiments based on 'metrics/mAP50(B)'
top_10_experiments = filtered_data.sort_values(
    by=[
        "metrics/mAP50-95(B)",
        "metrics/mAP50(B)",
        "val/box_loss",
        "val/cls_loss",
        "val/dfl_loss",
    ],
    ascending=False,
).head(15)

top_10_experiments
```

```python
import numpy as np

# Define the metrics and their ranking order (True for ascending, False for descending)
metrics_for_rrf = {
    "metrics/mAP50-95(B)": False,
    "metrics/mAP50(B)": False,
    "val/box_loss": True,
    "val/cls_loss": True,
    "val/dfl_loss": True,
}

# Constant k for RRF calculation
k = 60  # A common value, you can experiment with this

# Calculate ranks for each metric
for metric, ascending_order in metrics_for_rrf.items():
    rank_col_name = f"{metric}_rank"
    filtered_data[rank_col_name] = filtered_data[metric].rank(method='min', ascending=ascending_order)

# Calculate RRF score
rrf_score_col = 'rrf_score'
filtered_data[rrf_score_col] = 0
for metric in metrics_for_rrf:
    rank_col_name = f"{metric}_rank"
    filtered_data[rrf_score_col] += 1 / (k + filtered_data[rank_col_name])

# Sort by RRF score to find the best trials
top_trials_rrf = filtered_data.sort_values(by=rrf_score_col, ascending=False)

# Display the top 15 trials based on RRF
print("Top 15 trials based on RRF score:")
display(top_trials_rrf[[col for col in top_trials_rrf.columns if '_rank' not in col]].head(15))

# You can now use top_trials_rrf['trial_id'].head(15) for further steps
# like identifying the folders to keep.
```
