```python
import pandas as pd
import json

train = pd.read_csv("dataset/Train_df.csv")
```

```python
category_ids = train["class"].unique()
category_ids = {n: i for i, n in enumerate(sorted(category_ids))}

category_ids
```

```python
train.apply(
	lambda x: [x["xmin"], x["ymin"], x["xmax"] - x["xmin"], x["ymax"] - x["ymin"]], axis=1
)
```

```python
train["category_id"] = train["class"].map(category_ids)
train["id"] = train.index.values
train["bbox"] = train.apply(
	lambda x: json.dumps([x["xmin"], x["ymin"], x["xmax"] - x["xmin"], x["ymax"] - x["ymin"]]), axis=1
)
train["image_id"] = train["ImagePath"].values


train.sample(5)
```

```python
valid = pd.read_csv("dataset/Val_df.csv")

valid.sample(5)
```

```python
valid["category_id"] = valid["class"].map(category_ids)
valid["id"] = valid.index.values
valid["bbox"] = valid.apply(
	lambda x: json.dumps([x["xmin"], x["ymin"], x["xmax"] - x["xmin"], x["ymax"] - x["ymin"]]), axis=1
)
valid["image_id"] = valid["ImagePath"].values


valid.sample(5)
```

```python
valid["image_id"] = valid["image_id"].str.replace("train", "val")
```

```python
valid.sample(5)
```

```python
val_images = valid["image_id"].unique()
len(val_images)
```

```python
from PIL import Image, ExifTags
import cv2

for flag, v in ExifTags.TAGS.items():
    if v == "Orientation":
        break


def load_image(filepath):
    image = Image.open(filepath)

    exif = image._getexif()
    if exif is None:
        return image

    orientation_value = exif.get(flag, None)

    if orientation_value == 3:
        image = image.rotate(180, expand=True)
    elif orientation_value == 6:
        image = image.rotate(270, expand=True)
    elif orientation_value == 8:
        image = image.rotate(90, expand=True)
    return image


def load_image_cv2(filepath):
    return cv2.cvtColor(
        cv2.imread(filepath, cv2.IMREAD_IGNORE_ORIENTATION | cv2.IMREAD_COLOR),
        cv2.COLOR_BGR2RGB,
    )


for i in val_images:
    w1, h1 = Image.open(i).size
    w2, h2 = load_image(i).size
    img = load_image_cv2(i).shape
    w3, h3 = img[1], img[0]

    if w1 != w2 or h1 != h2 or h1 != h3 or w1 != w3:
        print(f"Image: {i}")
        print(f"Image size: {w1}, {h1}")
        print(f"Image size: {w2}, {h2}")
        print(f"Image size: {w3}, {h3}")

        print("\n")
```

```python
val_shapes = []
for i in val_images:
    w1, h1 = Image.open(i).size
    w2, h2 = load_image(i).size
    img = load_image_cv2(i).shape
    w3, h3 = img[1], img[0]
    val_shapes.append([i, w1, h1, w2, h2, w3, h3])

val_shapes = pd.DataFrame(val_shapes, columns=["image_id", "w1", "h1", "w2", "h2", "w3", "h3"])

val_shapes.sample(5)
```

```python
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor

train_shapes = []


def get_raw(i):
    w1, h1 = Image.open(i).size
    w2, h2 = load_image(i).size
    img = load_image_cv2(i).shape
    w3, h3 = img[1], img[0]
    return [i, w1, h1, w2, h2, w3, h3]


train_ids = train["ImagePath"].unique()
with ThreadPoolExecutor() as executor:
    train_shapes = list(tqdm(executor.map(get_raw, train_ids), total=len(train_ids)))

train_shapes = pd.DataFrame(
    train_shapes, columns=["image_id", "w1", "h1", "w2", "h2", "w3", "h3"]
)

train_shapes.sample(5)
```

```python
valid_merge = valid.merge(val_shapes, on="image_id",)
train_merge = train.merge(train_shapes, on="image_id",)
```
