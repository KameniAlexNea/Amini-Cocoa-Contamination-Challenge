```python
import pandas as pd
import json

train = pd.read_csv("dataset/Train_df.csv")
```

```python
train.sample(5)
```

```python
category_ids = train["class"].unique()
category_ids = {n: i for i, n in enumerate(sorted(category_ids))}

category_ids
```

```python
train.apply(
	lambda x: [x["xmin"], x["ymin"], x["xmax"] - x["xmin"], x["ymax"] - x["ymin"]], axis=1
)
```

```python
train["category_id"] = train["class"].map(category_ids)
train["id"] = train.index.values
train["bbox"] = train.apply(
	lambda x: json.dumps([x["xmin"], x["ymin"], x["xmax"] - x["xmin"], x["ymax"] - x["ymin"]]), axis=1
)
train["image_id"] = train["ImagePath"].values


train.sample(5)
```

```python
train.sample(5)
```

```python
valid = pd.read_csv("dataset/Val_df.csv")

valid.sample(5)
```

```python
valid["category_id"] = valid["class"].map(category_ids)
valid["id"] = valid.index.values
valid["bbox"] = valid.apply(
	lambda x: json.dumps([x["xmin"], x["ymin"], x["xmax"] - x["xmin"], x["ymax"] - x["ymin"]]), axis=1
)
valid["image_id"] = valid["ImagePath"].values


valid.sample(5)
```

```python
valid["image_id"] = valid["image_id"].str.replace("train", "val")
```

```python
valid.sample(5)
```

```python
columns = ["image_id", "bbox", "category_id", "id"]

train[columns].to_json(
	"dataset/train.json",
	# orient="records",
	# lines=True,
)
valid[columns].to_json(
	"dataset/valid.json",
	# orient="records",
	# lines=True,
)
```
